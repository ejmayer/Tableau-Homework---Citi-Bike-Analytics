```python
# dataframes = [jan_df, feb_df, mar_df, apr_df, may_df, jun_df] # july_df, aug_df, sept_df, oct_df, nov_df, dec_df]
dataframes = [jun_df, july_df, aug_df]
CitiBikeData_df = pd.concat(dataframes)
# CitiBikeData_df = jun_df
CitiBikeData_df.head()
```

```python
# reset index to show as rider id
CitiBikeData_df = CitiBikeData_df.rename(columns={'index': 'rideid'})
CitiBikeData_df.head()
```

```python
# recalculate and show trip duration as mins
CitiBikeData_df['Duration'] = CitiBikeData_df['tripduration']/60
CitiBikeData_df.Duration = CitiBikeData_df.Duration.round(2)
CitiBikeData_df.head()
```

```python
# use birth year to add Age column
CitiBikeData_df['Age'] = 2019 - CitiBikeData_df['birth year']
CitiBikeData_df.head()
```

```python
# Check max ages
CitiBikeData_df['Age'].nlargest(1000).value_counts()
```

```python
#Exclude ages above 90
CitiBikeData_df = CitiBikeData_df[CitiBikeData_df['Age'] < 90]
```

```python
# recheck max ages
CitiBikeData_df['Age'].nlargest(1000).value_counts()
```

```python
# create and restate gender of each rider in new gender column
CitiBikeData_df['Gender'] = CitiBikeData_df['gender'].replace([0,1,2],['unknown',"Male","Female"])
CitiBikeData_df = CitiBikeData_df.drop('gender',1)
CitiBikeData_df.head()
```
